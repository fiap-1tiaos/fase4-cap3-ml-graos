# Classificação de Grãos com Machine Learning

Este notebook tem como objetivo analisar o conjunto de dados de grãos de trigo e desenvolver modelos de classificação para identificar as variedades (Kama, Rosa, Canadian) com base em suas características físicas.

## 1. Análise e Pré-processamento de Dados

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações de visualização
sns.set(style="whitegrid")
%matplotlib inline
```

### Carregamento dos Dados
O dataset não possui cabeçalhos, então definiremos os nomes das colunas manualmente conforme a descrição da atividade.

```python
columns = ['Area', 'Perimeter', 'Compactness', 'Kernel Length', 'Kernel Width', 'Asymmetry Coefficient', 'Kernel Groove Length', 'Class']

# Usaremos sep='\s+' para lidar com múltiplos espaços/tabs.
df = pd.read_csv('../assets/seeds_dataset.txt', sep='\s+', names=columns, engine='python')

# Exibindo as primeiras linhas
df.head()
```

### Estatísticas Descritivas

```python
df.describe()
```

```python
df.info()
```

### Visualização da Distribuição

```python
# Histogramas para cada característica
df.hist(bins=20, figsize=(15, 10))
plt.suptitle('Distribuição das Características', fontsize=16)
plt.show()
```

```python
# Boxplots para verificar outliers
plt.figure(figsize=(15, 10))
sns.boxplot(data=df.drop(columns=['Class']), orient='h')
plt.title('Boxplot das Características')
plt.show()
```

### Análise de Correlação

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlação')
plt.show()
```

```python
# Pairplot para visualizar relações entre variáveis separadas por classe
sns.pairplot(df, hue='Class', palette='viridis')
plt.show()
```

### Verificação de Valores Ausentes

```python
missing_values = df.isnull().sum()
print("Valores ausentes por coluna:")
print(missing_values[missing_values > 0])

if missing_values.sum() == 0:
    print("\nNão há valores ausentes no conjunto de dados.")
```

## 2. Implementação e Comparação de Modelos

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Separando features e target
X = df.drop(columns=['Class'])
y = df['Class']

# Escalonamento dos dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Divisão em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
# Definindo os modelos
models = {
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "Random Forest": RandomForestClassifier(random_state=42)
}

results = {}

for name, model in models.items():
    # Treinamento
    model.fit(X_train, y_train)
    
    # Predição
    y_pred = model.predict(X_test)
    
    # Avaliação
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
    
    print(f"--- {name} ---")
    print(f"Acurácia: {acc:.4f}")
    print(classification_report(y_test, y_pred))
    
    # Matriz de Confusão
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Matriz de Confusão - {name}')
    plt.ylabel('Real')
    plt.xlabel('Previsto')
    plt.show()
```

```python
# Comparação dos Modelos
plt.figure(figsize=(10, 6))
sns.barplot(x=list(results.keys()), y=list(results.values()))
plt.title('Comparação da Acurácia dos Modelos')
plt.ylim(0.8, 1.0)
plt.ylabel('Acurácia')
plt.show()
```

## 3. Otimização de Modelos

Nesta etapa, utilizaremos o `GridSearchCV` para encontrar os melhores hiperparâmetros para os modelos, visando melhorar ainda mais o desempenho.

```python
from sklearn.model_selection import GridSearchCV

# Definindo os grids de parâmetros
param_grids = {
    "KNN": {
        'n_neighbors': [3, 5, 7, 9, 11],
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan']
    },
    "SVM": {
        'C': [0.1, 1, 10, 100],
        'kernel': ['linear', 'rbf', 'poly'],
        'gamma': ['scale', 'auto']
    },
    "Random Forest": {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10]
    }
}

optimized_results = {}
best_estimators = {}

for name, model in models.items():
    print(f"Otimizando {name}...")
    grid_search = GridSearchCV(estimator=model, param_grid=param_grids[name], cv=5, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X_train, y_train)
    
    best_model = grid_search.best_estimator_
    best_estimators[name] = best_model
    
    # Avaliação no conjunto de teste
    y_pred = best_model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    optimized_results[name] = acc
    
    print(f"Melhores Parâmetros para {name}: {grid_search.best_params_}")
    print(f"Acurácia Otimizada: {acc:.4f}")
    print("-" * 30)
```

```python
# Comparação Antes e Depois da Otimização
comparison_df = pd.DataFrame({
    'Modelo': list(results.keys()),
    'Original': list(results.values()),
    'Otimizado': list(optimized_results.values())
})

comparison_df = comparison_df.melt(id_vars='Modelo', var_name='Versão', value_name='Acurácia')

plt.figure(figsize=(12, 6))
sns.barplot(data=comparison_df, x='Modelo', y='Acurácia', hue='Versão')
plt.title('Comparação de Desempenho: Original vs Otimizado')
plt.ylim(0.8, 1.0)
plt.show()
```

## 4. Interpretação e Insights

Olhando para os resultados, deu pra perceber que todos os modelos mandaram muito bem, com acurácias bem altas. Isso provavelmente acontece porque as características físicas dos grãos (como área, perímetro, etc.) são bem distintas entre as variedades Kama, Rosa e Canadian. O gráfico de dispersão (pairplot) que a gente fez lá no começo já dava uma dica disso, mostrando grupos bem separadinhos.

Sobre a otimização com o Grid Search, confesso que eu esperava um salto gigante de desempenho, mas a melhoria foi pequena em alguns casos. Acho que isso mostra que os modelos com os parâmetros padrão do scikit-learn já são bem robustos para problemas "comportados" como esse. Mas valeu a pena testar para ter certeza que estamos extraindo o máximo possível dos dados.

O Random Forest e o SVM pareceram ser os mais consistentes. O KNN também foi bem, mas ele pode sofrer um pouco mais se a gente tivesse muito mais dados ou dimensões.

**Conclusão:**
Para automatizar a classificação, qualquer um desses modelos otimizados faria um trabalho excelente, muito mais rápido e provavelmente mais preciso que a classificação manual. Se eu tivesse que escolher um para colocar em produção, talvez fosse de Random Forest ou SVM pela robustez.
